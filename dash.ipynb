```python
import pandas as pd
from Capitulo import Capitulo
from Partida import Partida
from Subpartida import Subpartida
from Fraccion import Fraccion
```

```python
# Cambia 'ruta/al/archivo.xlsx' por la ruta de tu archivo
exp = pd.read_excel('/home/luis-loreto/Documentos/Trabajo/Bases/Cubo/Exportaciones/EUA/2024/CE Producto.xlsx')
imp = pd.read_excel('/home/luis-loreto/Documentos/Trabajo/Bases/Cubo/Importaciones/EUA/2024/CE Producto.xlsx')
```

```python
nuevas_columnas = exp['TIGIE'].str.split('---', expand=True)
exp.insert(0, 'Clave', nuevas_columnas[0])
exp.insert(1, 'Descripcion', nuevas_columnas[1])
exp.drop('TIGIE', axis=1, inplace=True)
nuevas_columnas = imp['TIGIE'].str.split('---', expand=True)
imp.insert(0, 'Clave', nuevas_columnas[0])
imp.insert(1, 'Descripcion', nuevas_columnas[1])
imp.drop('TIGIE', axis=1, inplace=True)
del nuevas_columnas
```

```python
imp['Capitulo'] = imp['Clave'].str[:2]
imp['Partida'] = imp['Clave'].str[:4]
imp['Subpartida'] = imp['Clave'].str[:6]
imp['Fraccion'] = imp['Clave']
```

```python
df_capitulo = pd.DataFrame(list(Capitulo.items()), columns=['A', 'Descripcion_capitulo'])
df_partida = pd.DataFrame(list(Partida.items()), columns=['A', 'Descripcion_partida'])
df_subpartida = pd.DataFrame(list(Subpartida.items()), columns=['A', 'Descripcion_subpartida'])
df_fraccion = pd.DataFrame(list(Fraccion.items()), columns=['A', 'Descripcion_fraccion'])
```

```python
cruce = pd.merge(pd.merge(pd.merge(imp, df_capitulo, left_on = imp['Capitulo'], right_on = 'A', how='inner'), 
                          df_partida, left_on = imp['Partida'], right_on = 'A', how = 'inner'), 
                          df_subpartida, left_on = imp['Subpartida'], right_on = 'A', how = 'inner')
```

```python
cruce = pd.merge(imp, df_capitulo, left_on = imp['Capitulo'], right_on = 'A', how='inner')[
    ['Clave', 'Capitulo', 'Descripcion_capitulo', 'Partida', 'Descripcion partida', 
     'Subpartida', 'Descripcion subpartida', 'Septiembre 2024']
    ]
```
